```python
import numpy as np
import torch
```

```python
#Numpy bridge
a = torch.ones(4, 3)
print(a)

b = a.numpy()
print(b)

a.add_(1)
print(b)
```

```python
a = np.ones(shape=[3, 4])
b = torch.from_numpy(a)
print(b)

np.add(a, 1, a)
print(a)
print(b)
```

```python
#Move the tensor to the cuda
c = torch.rand(3, 4)
print(c)

c = c.cuda()
print(c)
```

```python
CUDA = torch.cuda.is_available()

if CUDA:
    c = c.cuda()
    print(c)
```

```python
#Tensor processes
t1 = torch.rand(3, 4)
t2 = torch.rand(2, 4)

t_result1 = torch.concat([t1, t2], 0)
print(t_result1.shape)

t1 = torch.rand(4, 3)
t2 = torch.rand(4, 7)

t_result1 = torch.concat([t1, t2], 1)
print(t_result1.shape)
```

```python
#Adding dimension
i_tensor = torch.rand(3)
print(i_tensor)

tensor2 = i_tensor.unsqueeze(0)
print(tensor2, tensor2.size())

tensor3 = i_tensor.unsqueeze(1)
print(tensor3, tensor3.size())

tensor4 = tensor3.unsqueeze(1)
print(tensor4, tensor4.size())
tensor4 = tensor3.unsqueeze(2)
print(tensor4, tensor4.size())
```

```python
#Decrement dimension
i_tensor = torch.randn(2, 3, 4)[:, :, 0:2]
print(i_tensor)

tensor5 = i_tensor[:, :, 0:2]
print(tensor5, tensor5.size())

tensor5 = i_tensor[:, 0:2, 0]
print(tensor5, tensor5.size())

tensor5 = i_tensor[:, 0:0, 0:3]
print(tensor5, tensor5.size())
```
