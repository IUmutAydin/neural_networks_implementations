```python
import torch
import torch.nn as nn
```

```python
layer = nn.Linear(5, 5)
print(layer)
print(layer.weight.data)

nn.init.uniform_(layer.weight, 0.0, 3)
print(layer.weight.data)

nn.init.normal_(layer.weight)
print(layer.weight.data.mean(), layer.weight.data.std())
print(layer.weight.data)

nn.init.normal_(layer.weight, 0, 0.2)
print(layer.weight.data.mean(), layer.weight.data.std())
print(layer.weight.data)
```

```python
nn.init.constant_(layer.bias, 8)
print(layer.bias.data)

nn.init.zeros_(layer.bias)
print(layer.bias.data)
```

```python
nn.init.xavier_normal_(layer.weight)
print(layer.weight.data)
print(layer.weight.data.std())
```
