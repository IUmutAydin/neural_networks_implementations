```python
import torch
import torch.nn as nn
```

```python
#MSE Loss
prediction = torch.rand(3, 4)
label = torch.rand(3, 4)

loss1 = nn.MSELoss(reduction='none')

print((label - prediction)**2)
print(loss1(prediction, label))
print('\n')

loss2 = nn.MSELoss(reduction='mean')

print(((label - prediction)**2).mean())
print(loss2(prediction, label))
print('\n')

loss3 = nn.MSELoss(reduction='sum')

print(((label - prediction)**2).sum())
print(loss3(prediction, label))
```

```python
#BCE Loss
prediction = torch.rand(3, 4)
label = torch.zeros(3, 4).random_(0, 2)
sigmoid = nn.Sigmoid()
loss = nn.BCELoss()
logits = sigmoid(prediction)
print(loss(logits, label))

loss1 = nn.BCEWithLogitsLoss()
print(loss1(prediction, label))
```

```python
#Sparse CCE
prediction = torch.rand(3, 4)
label = torch.zeros(3, dtype=torch.long).random_(0, 4)
loss = nn.CrossEntropyLoss()

print(loss(prediction, label))
```
